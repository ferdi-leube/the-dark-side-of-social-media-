```python
licensesall = data[['Fiscal_Year','Inspections','Total_Licensees','Percent_Inspected']]
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(licensesall['Percent_Inspected'],model='additive', period=2)
decompose.plot()
plt.show()
```

```python
# indeed some seasonality, not perfectly statioary but much closer then developments we have, lets find a model
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(licensesall['Percent_Inspected'], lags=6)
```

auto correlated with 3 time lags in the past

```python
from statsmodels.tsa.stattools import adfuller
adft = adfuller(licensesall['Percent_Inspected'],autolag="AIC")
output_df = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used",
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_df)
```

```python
# lets make the model stationary
liclog = np.log(licensesall['Percent_Inspected'])
licdif = liclog.diff()
licdif = licdif.dropna()
plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(licdif)
plt.title('Stationary?')
plt.show()
```

```python
licensesalldif = licensesall.copy()
```

```python
licensesalldif['Percent_Inspected'] = licdif
```

```python
licensesalldif.dropna(inplace=True)
```

```python
split = round((len(licensesall)*0.8))
```

```python
train = licensesall['Percent_Inspected'][:split]
test = licensesall['Percent_Inspected'][split:]

from pmdarima.arima import auto_arima
model = auto_arima(train, trace=True, error_action='ignore', suppress_warnings=True, seasonal=False,  m=12)
res = model.fit(train)
```

```python
res.summary()
```

```python
from sklearn.metrics import r2_score
predtest = model.predict(n_periods=len(test))
predtest = pd.DataFrame(predtest,index = test.index,columns=['Prediction'])

r2_score(predtest,test)
```

```python
print(predtest)
print('\n')
print(test)
```

```python
plt.plot(predtest, color = "red")
plt.plot(test, color='black')
plt.title("Train/Test split for Passenger Data")
plt.ylabel("Passenger Number")
plt.xlabel('Year-Month')
sns.set()
plt.show()
```

```python
from autots import AutoTS, load_daily
```

```python
from autots import AutoTS, load_daily
model = AutoTS(
    forecast_length=21,
    frequency='infer',
    prediction_interval=0.9,
    ensemble=None,
    model_list="fast",  # "superfast", "default", "fast_parallel"
    transformer_list="fast",  # "superfast",
    drop_most_recent=1,
    max_generations=4,
    num_validations=2,
    validation_method="backwards"
)
```

```python
model = model.fit(
    licensesall,
    date_col='Fiscal_Year',
    value_col='Percent_Inspected'
    #id_col='series_id' if long else None
)
```

```python
prediction = model.predict()
```

```python
# result prediction, forecasted amount will lie somewhere in that interval

# lets prepare our finished dataset, for the predicted value I'll take the average between these intervals
```

```python
forecastdfnew2['Percent_Inspected_2'] = forecastdfnew['Percent_Inspected']
```

```python
forecastdfnew2['Average'] = (forecastdfnew2['Percent_Inspected'] + forecastdfnew2['Percent_Inspected_2'])/2
```

```python
finfore = pd.DataFrame({'Fiscal_Year':list(np.arange(2015,2019)),'Percent_Inspected':list(forecastdfnew2['Average'].values)[1:5]})
```

```python
finfore
```

```python
finlicenses = pd.concat([licensesall,finfore],axis=0,ignore_index=True)
```

```python
finlicenses.drop(['Inspections','Total_Licensees'],axis=1,inplace=True)
```

```python
finlicenses = finlicenses[finlicenses['Fiscal_Year']>=2002]
```
