```python
# lets look for outliers before finding mean to replace no infos with
def boxplots(data):
    import seaborn as sns
    import matplotlib.pyplot as plt
    numcol = list(data.describe().columns)
    for x in numcol:
        sns.boxplot(data = data, y=x)
        plt.title(x)
        plt.show()
```

```python
def number_of_ooutliers_per_column(data):
    
    numcol = list(data.describe().columns)
    for x in numcol:
        Q3 = data[x].quantile(0.75)
        Q1= data[x].quantile(0.25)
        IQR = Q3 - Q1
        filters = data[data[x]>Q3 + 1.5*IQR]
        print(x)
        print(len(filters))
        print('\n')
```

```python
# lets check the deelopment of number of gun violence crimes per year, and then update the 2018 value
incidents = all_wanted_years.groupby('Year').agg({'incident_id':'count'})
incidents.reset_index(inplace=True)
incidents
```
