```python
def boxplots(data):
    import seaborn as sns
    import matplotlib.pyplot as plt
    numcol = list(data.describe().columns)
    for x in numcol:
        sns.boxplot(data = data, y=x)
        plt.title(x)
        plt.show()
```

```python
boxplots(datamen)
```

```python
def number_of_ooutliers_per_column(data):
    
    numcol = list(data.describe().columns)
    for x in numcol:
        Q3 = data[x].quantile(0.75)
        Q1= data[x].quantile(0.25)
        IQR = Q3 - Q1
        filters = data[data[x]>Q3 + 1.5*IQR]
        print(x)
        print(len(filters))
        print('\n')
```

```python
number_of_ooutliers_per_column(datamen)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(datayear['Anxiety disorders (%)'],model='additive', period=2)
decompose.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(datayear['Anxiety disorders (%)'], lags=10)
```

```python
from statsmodels.tsa.stattools import adfuller
adft = adfuller(datayear['Anxiety disorders (%)'],autolag="AIC")
output_df = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used",
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_df)
```

```python
# lets try some TS models again, lets AutoTS, similar to TPOT going through different ARIMA models, checking multiple 
# statistics etc and even calcualtes lower and uppper limit of forecasts with best possible results
from autots import AutoTS
model = AutoTS(
    forecast_length=8,
    frequency='infer',
    prediction_interval=0.9,
    ensemble=None,
    model_list="fast",  # "superfast", "default", "fast_parallel"
    transformer_list="fast",  # "superfast",
    drop_most_recent=1,
    max_generations=4,
    num_validations=2,
    validation_method="backwards"
)
```
